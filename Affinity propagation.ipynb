```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("Mall_Customers.csv")
```

```python
dataset
```

```python
x=dataset.iloc[:,[3,4]].values
```

```python
x
```

```python
from sklearn.cluster import AffinityPropagation
```

```python
aff=AffinityPropagation(affinity='euclidean')
aff.fit(x)
plt.plot(x)
plt.show()
```

```python
y_means=aff.fit_predict(x)
```

```python
y_means
```

```python
supervised=(dataset)
```

```python
supervised['cluster_group']=y_means
```

```python
supervised
```

```python
supervised.to_csv("cluster.csv",index=False)
```

```python
centroids=aff.cluster_centers_
```

```python
centroids
```
